```python
def set_axis_style(ax, labels):
    ax.xaxis.set_tick_params(direction='out')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticklabels(labels,rotation=90)
    ax.set_xticks(np.arange(1, len(labels) + 1))
    ax.set_xlim(0.25, len(labels) + 1)
    
    
def plot_violin_graph(list_of_data, title):
    # Create a figure instance
    label = ["Basic", "Normal Predictor", "Baseline Only", "SVD", "SVDpp", "NMF", "SlopeOne", "CoClustering"]
    fig = plt.figure()
  
    # Create an axes instance
    ax = fig.gca()
    ax.set_palette="Set3"
    ax.set_title(title)
    set_axis_style(ax, label)
    # Create the violinplot
    violinplot = ax.violinplot(list_of_data)
    plt.show()
    
def plot_violin_individual_dataset(list_of_data, title):
    # Create a figure instance
    fig = plt.figure()
  
    # Create an axes instance
    ax = fig.gca()
    ax.set_title(title)
    # Create the violinplot
    violinplot = ax.violinplot(list_of_data)
    plt.show()
```

```python
def round_est_values(df):
    round_actual = np.round(df["est"], decimals=0)
    round_actual = round_actual.reset_index()
    df = df.reset_index()[["index", "actual", "details", "method"]]
    df = df.merge(round_actual, on="index", how="left")
    df = df.set_index("index")
    grouped = df.groupby("est", as_index=False).agg({"details":"count", "method":"last"})
    return grouped
```
